# **1.為什麼選擇權定價理論 Black Scholes 公式中,選擇權的價值與股票的預期報酬率​​μ​無關？直覺上而言,若公司的預期報酬率極高,則股票上漲的機率越高,那Call Option不是應該價值極高嗎？請說明如何合理化地解釋此現象。(注意,此處的”合理”可以請搜尋台灣各家券商發行權證的隱含波動率調整政策,請找出各家政策的共同之處,並說明產生此共同之處的原因。指現實中合理,也可以指在理論中合理,更甚至是以理論的缺陷說明其合理性)**

**Ans.**

布朗運動可以用來描述股票價格的隨機過程,由於布朗運動是連續的隨機過程, 處處不可微分, 無法使用古典微積分求解, 取而代之的是以隨機微分方程式來求解,這也可以稱做伊藤引理。

股票價格的走勢符合布朗運動, 這裡假設方程式是帶有飄移項和擴散項的幾何布朗運動, 假設$𝜇$(股票預期報酬率)與$𝜎$(隱含波動率)為常數

${𝒅}$$X(t)$ = $\mu$$\mathrm{𝒅}t$ + $𝜎$$\mathrm{𝒅}$ℬ(t)

股票的價格不能為負的, 不過股票的收益率可以是正的或負的, 這裡的$X(t)$可以當作股票的收益率

假設$S(t)$為股票的價格, ${𝒅}$$S(t)$為單位時間內價格的變化, ${𝒅}$$S(t)$/$S(t)$為股票的收益率, 替換${𝒅}$$X(t)$

${𝒅}$S(t)$ / $S(t)$ = $𝜇dt + 𝜎dℬ(t)

回到Black-Scholes微分方程, 其實此方程式已消除掉隨機的布朗運動項

令C為歐式看漲期權的價格,C為股票價格S和時間t的函數,表示為C(S,t)

對C使用伊藤引理

$𝒅C$ = ($\frac{\partial C}{\partial S}$𝑢S + $\frac{\partial C}{\partial t}$ + $1 \over 2$$\frac{\partial^2 C}{\partial S^2}$$σ^2$$S^2$)dt + $\frac{\partial C}{\partial S}$$σ$S$d$ℬ

令f為X(t)的二階連續可導函數並對t一階可導, f(X(t),t)

$𝒅f$ = $\frac{\partial f}{\partial t}$$𝒅t$ + $\frac{\partial f}{\partial X}$$𝒅X$ + $1 \over 2$$\frac{\partial^2 f}{\partial X^2}$$𝒅X^2$

假設f為以e為底的股票價格增長率, f = lnS 

對f使用隨機微分方程代數方法求解得到伊藤引理的一般形式:

$𝒅f$ = ($\frac{\partial f}{\partial t}$ + $\frac{\partial f}{\partial X}$$𝝖$ + $1 \over 2$$\frac{\partial^2 f}{\partial X^2}$$𝜷^2$)dt + $\frac{\partial f}{\partial X}$$𝜷$𝒅B

將S和C隨機微分方程離散化

$ΔC$ = ($\frac{\partial C}{\partial S}$𝑢S + $\frac{\partial C}{\partial t}$ + $1 \over 2$$\frac{\partial^2 C}{\partial S^2}$$σ^2$$S^2$)$Δ$t + $\frac{\partial C}{\partial S}$$σ$S$Δ$B

$ΔS(t)$ = $\mu$$Δ$t + $𝜎$S$Δ$B

由伊藤過程的X(t)函數與以e為底的股票價格增長率的隨機性來自於同一個布朗運動, 以此也可推得$Δ$C和$Δ$S來自於同一個布朗運動(因為$Δ$C由伊藤引理求解得), 因此我們可以用股票和期權建構一個對沖投資組合,設

[ -$𝗖$為放空一份期權(-1 x $𝗖$ = -$𝗖$),然後

做多+$\frac{\partial C}{\partial S}$份股票 ] 的對沖投資組合, 將股票和期權的權重$𝝖$和$𝞫$帶入$Δ$C和$Δ$S可以沖銷布朗運動ΔB,這種方法稱作Delta中性

設此投資組合的價值為P, 在$Δt$的變化為

$ΔP$ = (-$\frac{\partial C}{\partial t}$ - $1 \over 2$$\frac{\partial^2 C}{\partial S^2}$$σ^2$$S^2$)$Δ$t

這是一個無風險的投資組合, $ΔP$裡不存在布朗運動的隨機項$ΔB$,在$Δt$內的收益率必須等於無風險利率 $𝗿$,  $ΔP$ = $𝗿$$𝑷$$Δt$ , 將  $ ΔP$  和  $𝑷$ = -$𝗖$ + $\frac{\partial C}{\partial S}$S  代入式子做代數運算即可得到Black-Scholes微分方程式

$\frac{\partial C}{\partial t}$ + $𝗿$S$\frac{\partial C}{\partial S}$  - $1 \over 2$$σ^2S^2$$\frac{\partial^2 C}{\partial S^2}$ = $𝗿$$𝗖$

顯然股票的預期報酬率$𝜇$沒有出現在Black-Scholes中,也就是說在選擇權定價理論Black-Scholes公式中, 股票預期報酬率𝜇與選擇權價格$𝗖$無關, 而從布朗運動的性質, 服從B-S模型的股票價格是一個馬可夫過程, 當前的價格就已經包含對其未來做預測所需的全部資訊, 以馬可夫過程來觀察B-S式子可推論當前股票的價格$S$與期權的定價$𝗖$是互相影響著,而股票預期報酬率𝜇則受到股票當前內在價值所影響, B-S模型從理論上與現實中來說都是合理的。

# **2.請搜尋台灣各家券商發行權證的隱含波動率調整政策,請找出各家政策的共同之處,並說明產生此共同之處的原因。**

**Ans.** 

台灣各家券商對於發行權證的隱含波動率調整政策會依Vega風險做調整, 當波動率下滑時造成權證跌價的風險,權證是表示未來買進股票的權利,當權證對應的股票價格震盪激烈時,權證會越值錢, 因為股票漲越多,做多的人會賺越多, 如果股價下跌時只需付出權利金, 相反的, 當波動率震盪越趨越小, 震盪逐漸平穩, 持有部位的人手上的權證會變得不值錢,此所謂Vega風險。
 
基於此台灣各家券商的調整政策會傾向波動率向下,而不是向上,買賣權證的人會面臨著券商調降波動率,讓權證價格下跌的風險, 為了消除此風險,我們必須檢視券商發行的權證的公平性,選擇造市較好的券商。可以參考台灣權證商造市品質評比表來參考使用的券商。

# **3.請自行搜尋元大台灣50反向型ETF的官方網站,並瞭解其追蹤指數及成分、日常交易量,請試說明此ETF是否有更對投資者有吸引力的替代資產,若有,則為何此ETF尚未被其取代？**

**Ans.**


元大台灣50反型ETF適合在局勢動盪不安時投資, 50反型ETF顧名思義是跟著一般的50型ETF反向操作,一般的50型ETF是追縱台灣50指數的股票集合而成的基金, 裡面有許多台灣有影響力而且重要的公司, 如台積電,鴻海,聯發科,台達電,富邦金..., 一般型的台灣50型ETF擁有許多優點,分散風險,趨勢跟隨,透明度高,表現不佳的公司會被自然淘汰, ,因為這50間公司是有其價值的,以長期來看是有一定的投資報酬率,適合長期持有。

而元大台灣50反型ETF可以理解為跟一般型ETF反著做, 也就是說它在不正常情況下, 或著經濟週期在尤盛轉衰的時機, 又或者重大能影響經濟的新聞或事件時表現優異, 例如美國聯準會調高利率造成一波指數大行情股價下跌時表現優異, 又或者油價上漲造成美國經濟下滑而牽動台灣股票下挫的預期表現優異。 

元大台灣50反型ETF適合短期避險投資,投機,和投資組合的對沖, 在投資人有預期負面事件時投資避險,但不合長期持有,因為有著管理費過高,不配發股利的缺點,在市場不動盪趨於平穩時因為管理費過高它會慢慢的消耗投資人的資本, 股市大漲時又會造成損失,又長期來看因為貨幣供應的增加導致股市上升, 資本總有一天會被磨光,而當指數上升一倍,反向型ETF價值就會歸零,券商可能就宣布下架,其風險程度是高於一般的ETF,而一般的ETF其價值歸零的機率非常小。

反型ETF在投機,避險,對沖投資組合上有其優點,其性質類似於放空期權可以投機避險,當沖投資組合,來為投資人做風險控管, 降低突發事件與金融動盪的損失,也因為它也是使用指數50反做,所以相比放空期權,風險損失不會像放空期權那樣高, 當然如果是高風險偏好者選擇放空期權更佳, 而它在避險方面比放空期權更能分散風險部位和抓住熊市大行情時機點,因此在投資組合上還是有其地位, 並不容易被取代。
